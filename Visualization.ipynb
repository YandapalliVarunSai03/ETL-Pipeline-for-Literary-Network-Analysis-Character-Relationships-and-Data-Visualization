### Defining the function to get the transformed dataset

```python
def get_transformed_edge_data(client):
    # Access the database and collection
    db = client['DAP_Transformed_dataset']
    edges_collection = db['transformed_edges']

    # Retrieve the data from MongoDB and convert to DataFrame
    data = list(edges_collection.find({}, {"_id": 0}))  # Exclude MongoDB's internal _id field
    df = pd.DataFrame(data)
    
    print("\nData types of each column in transformed_edges:")
    print(df.dtypes)

    return df


def get_transformed_book_data(client):
    # Access the database and collection
    db = client['DAP_Transformed_dataset']
    books_collection = db['transformed_books']

    # Retrieve the data from MongoDB and convert to DataFrame
    data = list(books_collection.find({}, {"_id": 0}))  # Exclude MongoDB's internal _id field
    df = pd.DataFrame(data)

    print("\nData types of each column in transformed_books:")
    print(df.dtypes)
  
    return df


def get_transformed_characters_data(client):
    # Access the database and collection
    db = client['DAP_Transformed_dataset']
    characters_collection = db['transformed_characters']

    # Retrieve the data from MongoDB and convert to DataFrame
    data = list(characters_collection.find({}, {"_id": 0}))  # Exclude MongoDB's internal _id field
    df = pd.DataFrame(data)

    print("\nData types of each column in transformed_characters:")
    print(df.dtypes)

    return df
```

## Creating Character(Vertex) to Book(Edges) Mapping

```python
character_strength_df = edge_character_merged_df[edge_character_merged_df['connections_count']==6]
limited_df = character_strength_df.head(5)
```

```python
limited_df
```

```python
# As there are 19k rows in charcter table, we cannot make a dictonary for it to map the values 
# So we take only the id we want and then make a dictionary of only that ids

# Extracting unique character IDs from connected_character_ids
unique_ids = []
for ids in limited_df['connected_vertices']:
    for id in ids:
        unique_ids.append(id)
    
unique_book_ids = set(unique_ids)

# Filter the character df for these IDs
filtered_book_df = transformed_book_df[transformed_book_df['vertex_id'].isin(unique_book_ids)]

# Creating a dictionary mapping from filtered data
filtered_book_mapping = dict(zip(filtered_book_df['vertex_id'], filtered_book_df['book_names']))

# Display the filtered mapping
filtered_book_mapping
```

```python
# Map connected_character_ids to their corresponding names
limited_df['connected_vertices'] = limited_df['connected_vertices'].apply(
    lambda ids: [filtered_book_mapping.get(book_id, f"Unknown({book_id})") for book_id in ids]
)
```

```python
limited_df[['vertex_id', 'character_names', 'connected_vertices' ]]
```

```python
# Extract book names and their connected characters
edges = []
for index, row in limited_df.iterrows():
    character = row['character_names']
    books = row['connected_vertices']
    for book in books:
        edges.append((character, book))  # Created edges between books and characters

# Create the Graph
G = nx.Graph()
G.add_edges_from(edges)

# 2D Graph Visualization
plt.figure(figsize=(18, 12))  # Increased figure size for better visibility
pos = nx.spring_layout(G, seed=42)  # Spring layout for better spacing

# # Adjusting spring layout for tighter clusters
# pos = nx.spring_layout(G, seed=42, k=0.026, iterations=50)  # Smaller `k` for tighter spacing

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=800,
    node_color="skyblue",
    font_size=10,
    font_color="black",
    edge_color="blue"
)

plt.title("2D Graph Visualization of Characters and Books Appeared In", fontsize=15)
plt.show()
```

## Creating Book(Vertex) to Character(Edges) Mapping

```python
from collections import defaultdict
book_to_characters = defaultdict(list)

# Iterate through the DataFrame to map book IDs to character IDs
for _, row in transformed_edge_df.iterrows():
    character_id = row['vertex_id']
    for book_id in row['connected_vertices']:
        book_to_characters[book_id].append(character_id)

# Convert the defaultdict to a DataFrame
book_to_char_map = pd.DataFrame(
    {'book_id': list(book_to_characters.keys()), 
     'connected_character_ids': list(book_to_characters.values())}
)
book_to_char_map.shape
```

```python
book_to_char_map.head(10)
```

### Now mapping "book_id" with its "book names"

```python
book_to_char_map.rename(columns={'book_id': 'vertex_id'}, inplace=True)
book_to_char_map.info()
```

```python
# maping book id with book names

book_to_char_map=book_to_char_map.merge(transformed_book_df, how = "inner", on= "vertex_id")
book_to_char_map.info()
```

### Creating New Column for Count of Charcter Apperances per Book
